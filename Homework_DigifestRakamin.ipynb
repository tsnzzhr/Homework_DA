### Filtering Data

### Number of Finished and Canceled Transactions Each Year

```python
conditions = [
    (dframe['Invoice'].str.contains('A') == True),
    (dframe['Invoice'].str.contains('C') == True)]
choices = ['Adjust Debt', 'Cancelled']
dframe['Status'] = np.select(conditions, choices, default='Finished')
dframe
```

```python
dframe.groupby(['Status', 'Year'])['Invoice'].count()[2:]
```

### Cancellation Rate

```python
# cancel/total_invoice * 100%

cancel_amount = dframe.groupby(dframe['Status'] == 'Cancelled')["Invoice"].nunique()[1:]
total_invoice = dframe["Invoice"].nunique()

C_Rate = int(cancel_amount)/int(total_invoice)

print(f"{C_Rate:.0%}")

```
